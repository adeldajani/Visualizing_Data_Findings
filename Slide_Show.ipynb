```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('data_clean.csv')
```

## (What is the most common hour of the day that people ride bikes?)

> This visiualiztion will help us to understand at what times during the day were the Ford GoBike is used the most. it would be reasonable to assume the busiest hours are at the beginning and ending of work.

```python
plt.figure(figsize = (10,8))
base_color = sb.color_palette()[0]
sb.countplot(data = df, x = 'hour_of_day', color = base_color);
plt.xlabel('Hours of Day');
plt.ylabel('Number of Rides');
plt.title('How number of rides varies within the day');
```

As we assumed the 2 most common hours to start riding are at the beginning of work (8 am) and at the ending of work (5 pm)


## (What are the most popular start and end stations?)

> with the help of this vesualization we will be able to locate the most popular start point and end point

```python
plt.figure(1)
plt.title('Most Popular Start Stations', fontsize=15)
df.start_station_name.value_counts(ascending=True).tail(10).plot.barh();

plt.figure(2)
plt.title('Most Popular End Stations', fontsize=15)
df.end_station_name.value_counts(ascending=True).tail(10).plot.barh();
plt.show()
```

> 1) it is noted that the most popular start station is: Market St at 10th St

> 2) it is noted that the most popular end station is: San Francisco Caltrain Station 2 (Townsend St at 4th St)

## (What is the average trip duration every day for each gender?)

> Using this visualization we will compare the average trip duration every day for each gender

```python
plt.figure(figsize=(10,8))
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sb.stripplot(x = "day_of_week", y = "duration_min", hue = 'member_gender', data = df, order = day_order, jitter = 0.35, dodge = True, palette = "Dark2");
plt.xlabel('Day of Week');
plt.ylabel('Avg. Trip Duration in Minute');
plt.yscale('log');
plt.yticks([1, 2, 5, 10, 20, 50, 100], [1, 2, 5, 10, 20, 50, 100]);
ax = plt.gca()
ax.set_yticklabels([],minor = True);
plt.title("Avg. trip duration on every day of week for each gender");
```

We can note that Males have the highest trip durations everday compared to Females and Others
